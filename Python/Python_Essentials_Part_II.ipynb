## Logical Operators
The key logical operators are **and**, **or** & **not**

```python
x = True
y = False

print(f'Logical AND operation for {x} and {y}:- ',x  and  y)  #  True  if  both  values  are  true
print(f'Logical OR operation for {x} or {y}:- ',x  or  y)  #  True  if  either  of  the  values  is  true
# The above can be extended using multiple logical operators
print(f'NOT operation upon {y}:- ',not  y)  #  True  if  operand  is  false and vice versa
```

# if conditional statement

* The _if_ statement evaluates an expression and executes a block of code on the condition that the expression is true.
* In case the expression returns a False, the _if_ block-of-code is ignored.

```python
a=8
if a>6:
    print(a,"is greater than 6.") # Indentation for the block-of-code
print("Independent statement") # Not indented
```

Indentation refers to the spaces at the beginning of a
code line. It is the standard syntax in control statements where Python uses indentation
to indicate a block of code.

```python
age=21
if age>=18:
    print("The person is above the age limit.") # executed if the expression is True
    print("The present age meets the condition") # Multiple code lines may be part of the if block
```

#### else may be included to account for alternate statements when if condition is false

```python
x=5
if x==5:
    print('The if condition is true')
else: # when if condition is false
    print("The variable does not meet the condition")
    print('Else block-of-code got executed')
```

```python
x=-23
if x>0:
    print('The given number is positive.')
else:
    print('The number may be negative.')
print("Out of if-else blocks") # will be executed regardless of the if condition outcome
```

### Nested if
An _if_ condition can be said to be nested when it is part of another _if_ block-of-code.

The nested condition will only be tested when the parent _if_ expression happens to be true.

```python
i = 10
if (i >= 10): #  parent if statement

    if (i < 15): # Nested if statement
        print("i is smaller than 15")
    # These expressions are tested only when the parent if returns True.
    if (i < 12):
        print("i is smaller than 12 too")
    else:
        print("i is greater than 15")
```

```python
# bp range 120-129 accepted as normal
bp_systolic = int(input("Enter the current bp measurement:"))
condition = 120<=bp_systolic<=129
if condition == False: # out of normal range
    print("The blood pressure is out of the normal range.")
    if bp_systolic<90:
        print("A case of Low blood pressure")
        print("Advised to stay hydrated with regular intake of fluids")
    elif bp_systolic>=130:
        print("A case of high blood pressure")
        print("30 minutes of daily walk is recommended.")
    else:
        print("Not necessarily low bp but fluid intake to be increased")
else:
    print("Blood pressure is in the normal range")
```

### Set functions

```python
news = {"Business","Politics"}
news
```

```python
news.add("Sports") # addition of a single item
print(news)
```

```python
sections = ["Entertainment","Lifestyle","Advertising"]
news.update(sections)
news # mutiple items added
```

```python
news.add(250)
print(news)
news.remove(250)
news.discard("Lifestyle")
print("-------")
print(news)
```

#### While we can modify the set, a given set item is not replaceable. In other words, sets are mutable but set items are not.

```python
news.clear() # Delete all items within the set
news
```

```python
del news # Delete the set object
print(news)
```
