# Indexing and Selection 

```python
import numpy as np
```

```python
array = np.arange(10,200,9)
array
```

```python
array[8]
```

```python
array[ 4:  9: 2 ] #(start default=index 0 , Stop default= last index  , Jump/step size )
```

```python
print(array)
np.where(array >= 100) # where ever your condition is True it will return its Index
```

Indexing in 2D

```python
arr_2d =np.arange(1,10).reshape(3,3) 
arr_2d
```

```python
arr_2d[0] # all the values in 0th row
```

```python
arr_2d[1 : 2, 1 : 3] # [ Start row : End row , Start Column : End Column ] Note End {row, column} are excluded 
```

```python
arr_2d[2: , : 2 ]
```

Fancy Indexing (aka Advance Indexing)

```python
arr_2d
```

```python
rows = [1,1,0,2,2,0]
column = [0,2,2,1,0,1]

arr_2d[rows, column]
```
